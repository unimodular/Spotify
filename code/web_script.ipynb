```python
import csv
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# 创建 Spotify 客户端实例
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="cd38960ac8cb4e6387ec9d2214b84f6f",
                                                           client_secret="67dabe2dd4064512b16294aa0c2f78b1"))

# 搜索关键词 'Taylor Swift'，同时获取 tracks 和 episodes
results = sp.search(q='Taylor Swift', type='track,episode', limit=20)

# 提取信息并组织数据
data = []

# 处理 tracks
for track in results['tracks']['items']:
    data.append({
        'Type': 'Track',
        'Name': track['name'],
        'Description': ', '.join(artist['name'] for artist in track['artists']),
    })

# 处理 episodes
for episode in results['episodes']['items']:
    data.append({
        'Type': 'Episode',
        'Name': episode['name'],
        'Description': episode['description'],
    })

# 保存为 CSV 文件
csv_file = 'spotify_results.csv'
csv_columns = ['Type', 'Name', 'Description']

try:
    with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=csv_columns)
        writer.writeheader()
        writer.writerows(data)
    print(f"数据已保存到 {csv_file}")
except Exception as e:
    print(f"保存文件时出错: {e}")
```

```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# 创建 Spotify 客户端实例
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="cd38960ac8cb4e6387ec9d2214b84f6f",
                                                           client_secret="67dabe2dd4064512b16294aa0c2f78b1"))

# 搜索包含 podcast 和 show 的结果
query = "Taylor Swift"  # 替换为您的关键词
results = sp.search(q=query, type='show,episode', limit=20)

# 提取 episodes 信息
data = []
if 'episodes' in results:
    for episode in results['episodes']['items']:
        data.append({
            'name_episode': episode['name'],
            'description': episode['description'],
            'release_date': episode['release_date'],
            'duration_ms': episode['duration_ms'],
            'id': episode['id']
        })

# 提取 shows 信息（如果需要）
if 'shows' in results:
    for show in results['shows']['items']:
        data.append({
            'name_episode': show['name'],
            'description': show['description'],
            'release_date': show.get('release_date', 'N/A'),  # Shows may not have release dates
            'duration_ms': 'N/A',  # Shows don't have a fixed duration
            'id': show['id']
        })

# 创建 Pandas DataFrame
df = pd.DataFrame(data)

# 导出为 CSV 文件
output_file = "spotify_episodes_shows.csv"
df.to_csv(output_file, index=False, encoding='utf-8')

# 打印预览
print("数据预览：")
print(df.head())

print(f"数据已保存到 {output_file}")
```

```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# 创建 Spotify 客户端实例
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="cd38960ac8cb4e6387ec9d2214b84f6f",
                                                           client_secret="67dabe2dd4064512b16294aa0c2f78b1"))

# 搜索包含 podcast episodes 的结果
query = "Taylor Swift"  # 替换为您的关键词
results = sp.search(q=query, type='episode', limit=50)  # 搜索类型为 episode

# 提取 podcast episodes 信息
data = []
for episode in results['episodes']['items']:
    data.append({
        'name_episode': episode['name'],          # 单集名称
        'description': episode['description'],   # 单集描述
        'release_date': episode['release_date'], # 发布日期
        'duration_ms': episode['duration_ms'],   # 时长（毫秒）
        'id': episode['id']                      # 唯一标识符
    })

# 创建 Pandas DataFrame
df = pd.DataFrame(data)

# 保存为 CSV 文件
output_file = "podcast_episodes_taylor_swift.csv"
df.to_csv(output_file, index=False, encoding='utf-8')

# 打印预览
print("数据预览：")
print(df.head())

print(f"数据已保存到 {output_file}")
```
